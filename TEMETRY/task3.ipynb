```python
#. During a CubeSat flight, the avionics computer records telemetry
#parameters for the payload(CubeSat) until it abruptly stops.

import matplotlib.pyplot as plt

T = 50000   
m = 1000    
g = 9.81    
dt = 1      
t_total = 16

time = []
velocity = []
altitude = []

v = 0  
h = 0  

for t in range(0, t_total + 1):
    if t <= 10:   
        a = T/m - g
    else:        
        a = -g
    
    v = v + a*dt
    h = h + v*dt

    time.append(t)
    velocity.append(v)
    altitude.append(h)


max_v = max(velocity)
max_h = max(altitude)
t_vmax = time[velocity.index(max_v)]
t_hmax = time[altitude.index(max_h)]


fig, ax = plt.subplots(2, 1, figsize=(8, 6))
fig.suptitle("CubeSat Flight Telemetry Simulation")


ax[0].plot(time, altitude, 'b', label="Altitude (m)")
ax[0].scatter(t_hmax, max_h, color='r', label="Max Altitude")
ax[0].set_ylabel("Altitude (m)")

ax[1].plot(time, velocity, 'g', label="Velocity (m/s)")
ax[1].scatter(t_vmax, max_v, color='r', label="Max Velocity")
ax[1].set_ylabel("Velocity (m/s)")
ax[1].set_xlabel("Time (s)")


plt.tight_layout()
plt.show()
```
