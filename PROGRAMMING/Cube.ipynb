```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 100, 1000)

station1 = 30 * np.sin(0.1 * t) + 10 
station2 = 30 * np.sin(0.1 * t - np.pi/2) + 10 

handover_times = []

for i in range(1, len(t)):
    if station1[i-1] > 0 and station1[i] <= 0 and station2[i] > 0:
        handover_times.append(t[i])

plt.figure(figsize=(9, 6))

plt.plot(t, station1, 'r')
plt.plot(t, station2, 'b')

plt.axhline(0, color='k', linestyle='--')

for x in handover_times:
    plt.axvline(x, color='r', linestyle='--', alpha=0.7)
    plt.text(x, 20, f"Handover at {x:.1f} min", rotation=90, color='black', fontsize=10)
plt.title("CubeSat Ground Station Handover (Simple Version)")
plt.xlabel("Time (minutes)")
plt.ylabel("Elevation (degrees)")
plt.show()
```
