```python
import matplotlib.pyplot as plt
import random
import time
from IPython.display import clear_output

times = []
temp = []
volt = []
curr = []


for i in range(50):
   
    t = i * 0.5 
    
    temperature = random.uniform(25, 35)   
    voltage = random.uniform(3.3, 3.8)     
    current = random.uniform(0.4, 0.9)     

    times.append(t)
    temp.append(temperature)
    volt.append(voltage)
    curr.append(current)

    clear_output(wait=True)

    fig_1 = plt.figure(figsize=(8, 6))
    fig_1.suptitle("Real Time Telemetry Updates", fontsize=14, fontweight='bold')

    ax1 = fig_1.add_subplot(3, 1, 1)
    ax2 = fig_1.add_subplot(3, 1, 2)
    ax3 = fig_1.add_subplot(3, 1, 3)

    ax1.plot(times, temp, 'r')
    ax1.set_ylabel("Temp (C)")
   

    ax2.plot(times, volt, 'b')
    ax2.set_ylabel("Voltage (V)")

    ax3.plot(times, curr, 'g')
    ax3.set_ylabel("Current (A)")
    ax3.set_xlabel("Time (s)")
   
    plt.tight_layout()
    plt.show()

    time.sleep(0.2)
```
